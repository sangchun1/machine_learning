```python
from sklearn.datasets import make_regression
# 가상의 회귀분석 데이터 생성 함수
# n_features : 독립변수의 갯수, bias: y절편
# noise: y의 표준변차
# coef: 회귀계수(기울기) 사용여부, True이면 w값을 리턴함
X, y, w = make_regression(
    n_samples=50, n_features=1, bias=100, noise=10, random_state=0, coef=True
)
print(X[:5].flatten())
print(y[:5])
print(w)
```

```python
import numpy as np
xx = np.linspace(-3, 3, 100) # -3~3, 100등분
y0 = w * xx + 100
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(xx, y0, "r-")
plt.scatter(X, y, s=100)
plt.xlabel("x")
plt.ylabel("y")
plt.title("regression")
```
