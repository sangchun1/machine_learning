```python
import pandas as pd

df = pd.read_csv("c:/data/iris/iris.csv")
df.columns
```

```python
X = df[df.columns[:-2]]
y = df['Species']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
y_score = model.decision_function(X_test)
print(y_score[::20])
print(model.predict(X_test[::20]))
```

```python
from sklearn.metrics import confusion_matrix

y_hat = model.predict(X_test)
confusion_matrix(y_test, y_hat)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_hat))
```

```python
# roc_curve() 함수는 이진분류에만 지원됨
from sklearn.metrics import roc_curve
# fpr, tpr, thresholds = roc_curve(y, model.decision_function(X))
# fpr, tpr, thresholds
# 다중 클래스에 대해서는 ROC Curve를 그릴 수 없으므로
# 각각의 클래스에 대해 별도로 ROC Curve를 그려야 함
# y변수를 원핫인코딩한 리스트
from sklearn.preprocessing import label_binarize

y = label_binarize(y, classes=[0,1,2])
y[::20]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=0)
```

```python
%matplotlib inline
from sklearn.metrics import auc
import matplotlib.pyplot as plt

n_classes = 3
fpr = dict()
tpr = dict()
roc_auc = dict()
y_score = model.decision_function(X_test)
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
print(roc_auc)
for i in range(n_classes):
    plt.figure()
    plt.plot(fpr[i], tpr[i], label='ROC curve (area = %.3f)' % roc_auc[i])
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([-0.05, 1.05])
    plt.ylim([-0.05, 1.05])
    plt.xlabel('FPR')
    plt.ylabel('TPR')
    plt.title('ROC')
    plt.legend(loc="lower right")
    plt.show()
```
