```python
import datetime
import cv2
capture = cv2.VideoCapture("c:/data/video/duck.mp4")
#인코딩 방식 설정(코덱)
fourcc = cv2.VideoWriter_fourcc(*'XVID')
record = False
while True:
    if(capture.get(cv2.CAP_PROP_POS_FRAMES) == capture.get(cv2.CAP_PROP_FRAME_COUNT)):
        capture.open("c:/data/video/duck.mp4")
    ret, frame = capture.read()
    cv2.imshow("VideoFrame", frame)
    #날짜_시간-분-초
    now = datetime.datetime.now().strftime("%d_%H-%M-%S")
    key = cv2.waitKey(33)
    if key == 27: #ESC
        break
    elif key == 26: # Ctrl Z
        print("캡쳐")
        cv2.imwrite("c:/data/images/capture"+now+".png", frame)
    elif key == 24: # Ctrl X
        print("녹화 시작")
        record = True
        video = cv2.VideoWriter("c:/data/images/capture"+now+".avi", fourcc, 20.0, (frame.shape[1], frame.shape[0]))
    elif key == 3: # Ctrl C
        print("녹화 중지")
        record = False
        video.release()

    if record == True:
        print("녹화 중..")
        video.write(frame)
capture.release()
```
